```python
dna = "AGGTCAATGCATTCCAAAATTGGGGCATGACTGCTGACCAGGGTGATGCATGCATGCTGACTGACGTCCT"
```

```python
dna_length = len(dna)
print(dna_length)
```

```python
range(dna_length)
```

```python
reversed_dna = dna[::-1]
print(reversed_dna)
```

```python
complement = ""
for i in reversed_dna:
  if i == "A":
    complement += "T"
  elif i == "T":
    complement += "A"
  elif i == "G":
    complement += "C"
  elif i == "C":
    complement += "G"
print("Reverse Complement: ", complement)
```
