```python
import docx
```

```python
file_name = 'Mus musculus.docx'
```

```python
def get_text_from_docx(file_path):
  doc = docx.Document(file_path)
  full_text = []
  for para in doc.paragraphs:
    full_text.append(para.text)
  return '\n'.join(full_text)
```

```python
def calculate_at_content(sequence):
  seq = sequence.upper().replace(' ', '').replace('\n', '')
  a_count = seq.count('A')
  t_count = seq.count('T')
  total = len(seq)
  if total == 0:
    return 0
  at_content = ((a_count + t_count) / total) * 100
  return at_content
```

```python
extracted_text = get_text_from_docx(file_name)
print("Extracted text from word file :\n", extracted_text)
```

```python
dna_sequence = extracted_text.strip()
```

```python
at_content = calculate_at_content(dna_sequence)
print(f"AT content: {at_content:.2f}%")
```

```python
def calculate_pyrimidine_content(sequence):
  seq = sequence.upper().replace(' ','').replace('\n','')
  t_count = seq.count('T')
  c_count = seq.count('C')
  total = len(seq)
  if total == 0:
    return 0
  pyrimidine_content = ((t_count + c_count) / total) * 100
  return pyrimidine_content
```

```python
pyrimidine_content = calculate_pyrimidine_content(dna_sequence)
print(f"Pyrimidine content: {pyrimidine_content:.2f}%")
```
